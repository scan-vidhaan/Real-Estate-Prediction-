```python
import pandas as pd
url='Bengaluru_House_Data.csv'
dataset=pd.read_csv(url)
dataset.head(30)
```

```python
dataset = dataset.drop(columns=['society'])
dataset.head()
```

```python
dataset["availability"] = dataset["availability"].apply(lambda x: 1 if x == "Ready To Move" else 0)
dataset.head()
```

```python
dataset.head(30)
```

```python
dataset["balcony"] = dataset["balcony"].apply(lambda x: 0 if pd.isna(x) else x)
dataset.head(30)
```

```python
dataset['total_sqft'] = pd.to_numeric(dataset['total_sqft'], errors='coerce')
dataset['price_per_sqft']=(dataset['price']*100000)/dataset['total_sqft']
dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sqft_price=dataset['price_per_sqft']
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

dataset=remove_outliers(dataset,"price_per_sqft")

plt.figure(figsize=(12, 6))
sns.boxplot(x=dataset['price_per_sqft'])
plt.title("Price Per Square Foot Distribution in Bangalore")
plt.xlabel("Price per Sqft (INR)")
plt.show()
```

```python
dataset.head()
```

```python
dataset["size"] = dataset["size"].str.extract("(\d+)").astype(float)
dataset.head()
```

```python
# Feature Engineering from price_per_sqft (without directly using it)
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Step 1: Location-wise median price per sqft
location_psf_median = dataset.groupby('location')['price_per_sqft'].median()
dataset['location_median_psf'] = dataset['location'].map(location_psf_median)

# Step 2: Difference between property psf and location median psf
dataset['psf_diff_from_location'] = dataset['price_per_sqft'] - dataset['location_median_psf']

# Step 3: Categorical bucketing of price_per_sqft (low/mid/high)
q1 = dataset['price_per_sqft'].quantile(0.25)
q3 = dataset['price_per_sqft'].quantile(0.75)

def psf_category(x):
    if x <= q1:
        return 'low'
    elif x <= q3:
        return 'mid'
    else:
        return 'high'

dataset['psf_category'] = dataset['price_per_sqft'].apply(psf_category)

# Encode the category
le = LabelEncoder()
dataset['psf_category'] = le.fit_transform(dataset['psf_category'])

# Step 4: Drop original price_per_sqft to avoid leakage
dataset.drop(columns=['price_per_sqft'], inplace=True)

# Final train-test split
X = dataset.drop(columns=['price'])
y = dataset['price']

print("Features ready for modeling. X shape:", X.shape)
```

```python
dataset['psf_category'].head(20)
```

```python
dataset = pd.get_dummies(dataset, columns=["area_type","location",], drop_first=True)


dataset.head()
```

```python
# Select one-hot encoded area_type and location columns
cols_to_apply = dataset.filter(regex="^area_type_|^location_").columns

# Fill any NaNs in those columns with 0 before converting to int
dataset[cols_to_apply] = dataset[cols_to_apply].fillna(0).astype(int)

# Drop any remaining rows with NaNs in the entire dataset
dataset = dataset.dropna()
count=dataset.filter(regex="^location_").sum(axis=1).value_counts()
print(count)

# Confirm it's clean
dataset.head(30)

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize model
decision_tree_model = DecisionTreeRegressor()


# Separate x and y again


# Prepare data
x = dataset.drop(columns='price')

y = dataset['price']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train model
decision_tree_model.fit(X_train, y_train)

# Predict
y_pred = decision_tree_model.predict(X_test)

# Compute metrics
dt_mae = mean_absolute_error(y_test, y_pred)
dt_mse = mean_squared_error(y_test, y_pred)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, y_pred)

# Print metrics
print("Mean Absolute Error (MAE):", dt_mae)
print("Mean Squared Error (MSE):", dt_mse)
print("Root Mean Squared Error (RMSE):", dt_rmse)
print(f"R-squared Score (R): {dt_r2:.4f}")

# Provide verdicts based on metric values
if dt_mae < 2:
    print("Verdict: The model has a low MAE, indicating high accuracy in predicting property prices.")
else:
    print("Verdict: The MAE is relatively high, suggesting the model may need improvement.")

if dt_rmse < 150:
    print("Verdict: The RMSE is within an acceptable range, implying low prediction errors.")
else:
    print("Verdict: The RMSE is high, indicating significant deviations in predictions.")

if dt_r2 > 0.90:
    print("Verdict: The model explains most of the variance in the data, which is excellent.")
elif dt_r2 > 0.75:
    print("Verdict: The model performs well but can be further improved.")
else:
    print("Verdict: The model's R score is low, indicating room for significant improvement.")
```
