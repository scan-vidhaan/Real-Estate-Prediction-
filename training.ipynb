```python
import pandas as pd
url='Bengaluru_House_Data.csv'
dataset=pd.read_csv(url)
dataset.head(30)
```

```python
dataset = dataset.drop(columns=['society'])
dataset.head()
```

```python
dataset["availability"] = dataset["availability"].apply(lambda x: 1 if x == "Ready To Move" else 0)
dataset.head()
```

```python
dataset.head(30)
```

```python
dataset["balcony"] = dataset["balcony"].apply(lambda x: 0 if pd.isna(x) else x)
dataset.head(30)
```

```python
dataset['total_sqft'] = pd.to_numeric(dataset['total_sqft'], errors='coerce')
dataset['price_per_sqft']=(dataset['price']*100000)/dataset['total_sqft']
dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sqft_price=dataset['price_per_sqft']
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

dataset=remove_outliers(dataset,"price_per_sqft")

plt.figure(figsize=(12, 6))
sns.boxplot(x=dataset['price_per_sqft'])
plt.title("Price Per Square Foot Distribution in Bangalore")
plt.xlabel("Price per Sqft (INR)")
plt.show()
```

```python
dataset.head()
```

```python
dataset["size"] = dataset["size"].str.extract("(\d+)").astype(float)
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns=["area_type","location",], drop_first=True)


dataset.head()
```

```python
cols_to_apply = dataset.filter(regex="^area_type_|^location_").columns

dataset[cols_to_apply] = dataset[cols_to_apply].astype(int) 
# Drop all rows with any NaN values
dataset = dataset.dropna()

dataset.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize model
decision_tree_model = DecisionTreeRegressor()


# Separate x and y again


# Prepare data
x = dataset.drop(columns='price')
y = dataset['price']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train model
decision_tree_model.fit(X_train, y_train)

# Predict
y_pred = decision_tree_model.predict(X_test)

# Compute metrics
dt_mae = mean_absolute_error(y_test, y_pred)
dt_mse = mean_squared_error(y_test, y_pred)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, y_pred)

# Print metrics
print("Mean Absolute Error (MAE):", dt_mae)
print("Mean Squared Error (MSE):", dt_mse)
print("Root Mean Squared Error (RMSE):", dt_rmse)
print(f"R-squared Score (R): {dt_r2:.4f}")

# Provide verdicts based on metric values
if dt_mae < 2:
    print("Verdict: The model has a low MAE, indicating high accuracy in predicting property prices.")
else:
    print("Verdict: The MAE is relatively high, suggesting the model may need improvement.")

if dt_rmse < 150:
    print("Verdict: The RMSE is within an acceptable range, implying low prediction errors.")
else:
    print("Verdict: The RMSE is high, indicating significant deviations in predictions.")

if dt_r2 > 0.90:
    print("Verdict: The model explains most of the variance in the data, which is excellent.")
elif dt_r2 > 0.75:
    print("Verdict: The model performs well but can be further improved.")
else:
    print("Verdict: The model's R score is low, indicating room for significant improvement.")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Assuming your dataset is preprocessed and stored in 'dataset'
x = dataset.drop(columns='price')
y = dataset['price']

# Split data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Hyperparameter grid
param_grid = {
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

# Initialize model and perform grid search
dt = DecisionTreeRegressor(random_state=42)
grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best model
best_dt = grid_search.best_estimator_
y_pred = best_dt.predict(X_test)

# Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Output results
print("Best Hyperparameters:", grid_search.best_params_)
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared Score (R): {r2:.4f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor

param_dist = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [4, 6, 8],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 1, 2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'reg_lambda': [0.5, 1, 1.5]
}

xgb = XGBRegressor(objective='reg:squarederror', random_state=42)
rand_search = RandomizedSearchCV(xgb, param_distributions=param_dist, 
                                 n_iter=30, scoring='neg_mean_squared_error', 
                                 cv=3, verbose=1, n_jobs=-1)
rand_search.fit(X_train, y_train)

best_model = rand_search.best_estimator_
y_pred_best = best_model.predict(X_test)

# Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
new_mae = mean_absolute_error(y_test, y_pred_best)
new_mse = mean_squared_error(y_test, y_pred_best)
new_rmse = np.sqrt(mse)
new_r2 = r2_score(y_test, y_pred_best)

print(" Improved XGBoost with Randomized Search:")
print(f"MAE: {new_mae:.4f} | MSE: {new_mse:.4f} | RMSE: {new_rmse:.4f} | R: {r2:.4f}")
```
